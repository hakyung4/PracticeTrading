```python
#convert data to dataframe
aapl = close.AAPL.copy().to_frame()
aapl
```

```python
#shift down data by one
aapl["lag1"] = aapl.shift(periods=1)
aapl
```

```python
#By shifting down data by one like above, we can calculate the percentage change of daily return
```

```python
aapl["Diff"] = aapl.AAPL.sub(aapl.lag1)
aapl
```

```python
aapl["% Change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl
```

```python
#another method to do the same thing
aapl["Diff2"] = aapl.AAPL.diff(periods=1)
aapl
```

```python
aapl["% Change 2"] = aapl.AAPL.pct_change(periods=1).mul(100)
aapl
```

```python
del aapl["Diff2"]
del aapl["% Change 2"]
```

```python
ret = aapl.pct_change().dropna()
ret
```

```python
ret.info()
```

```python
ret.plot(kind="hist", figsize=(12,8), bins=100)
plt.show()
```

```python
daily_mean_ret = ret.mean()
print(daily_mean_ret)
var_daily = ret.var()
print(var_daily)
```

```python
std_daily = np.sqrt(var_daily)
print(std_daily)
```

```python
annual_mean_ret = daily_mean_ret * 252
annual_mean_ret
```

```python
annual_var_ret = var_daily * 252
print(annual_var_ret)
annual_std_ret = np.sqrt(annual_var_ret)
print(annual_std_ret)
```

```python
ticket = ["SPY", "AAPL", "KO", "IBM", "DIS", "MSFT"]
```

```python
stocks = yf.download(ticket, start = "2010-01-01", end = "2021-01-01")
```

```python
close = stocks.loc[:,"Close"].copy()
```

```python
norm_close = close.div(close.iloc[0]).mul(100)
```

```python
norm_close.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret = close.pct_change().dropna()
ret.head()
```

```python
ret.describe()
```

```python
ret.describe().T
```

```python
summary = ret.describe().T.loc[:,["mean", "std"]]
summary
```

```python
summary["mean"] = summary["mean"] * 252
summary["std"] = summary["std"] * np.sqrt(252)
summary
```

```python
summary.plot.scatter(x = "std", y = "mean", figsize=(12,8), s = 50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy = (summary.loc[i, "std"] + 0.002, summary.loc[i, "mean"] + 0.002), size = 15)

plt.xlabel("Annual risk(std)", fontsize = 15)
plt.ylabel("Annual return(mean)", fontsize = 15)
plt.title("Risk/Return", fontsize = 25)
plt.show()
```
