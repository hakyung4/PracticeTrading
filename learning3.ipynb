### Cumulative returns, drawdowns, etc

```python
apple = yf.download("AAPL")
```

```python
apple = apple.Close.to_frame()
```

```python
apple
```

```python
apple.Close
```

```python
apple["daily_ret"] = np.log(apple.div(apple.shift(1)))
apple
```

```python
apple.dropna(inplace=True)
apple
```

```python
apple.daily_ret.sum()
```

```python
# what if I invested $ 1?
np.exp(apple.daily_ret.sum())
```

```python
apple["cumulative_ret"] = apple.daily_ret.cumsum().apply(np.exp)
apple
```

```python
apple.cumulative_ret.plot(figsize=(15,8), title="Apple buy and hold", fontsize=12)
plt.show()    
```

```python
# annual avg return
apple.daily_ret.mean() * 252
```

```python
# annual std
apple.daily_ret.std() * np.sqrt(252)
```

```python
# Calculate drawdowns
# drawdown: how much an investment or trading account is down from the peak before it recovers back to the peak.
apple["cumulative_max"] = apple.cumulative_ret.cummax()
apple
```

```python
apple[["cumulative_ret", "cumulative_max"]].plot(figsize=(15,8), title="Apple buy and hold + cumulative max", fontsize=12)
plt.show()
```

```python
apple["drawdown"] = apple["cumulative_max"] - apple["cumulative_ret"]
apple
```

```python
apple.drawdown.max()
```

```python
apple.drawdown.idxmax()
```

```python
apple.loc[(apple.index=='2022-05-19')]
```

```python
apple.loc[(apple.index<='2022-05-19')]
```

```python
apple["drawdown %"] = (apple["cumulative_max"] - apple["cumulative_ret"]) / apple["cumulative_max"]
apple
```

```python
apple["drawdown %"].max()
```

```python
apple["drawdown %"].idxmax()
```

```python
# SMA (Simple Moving Average) Strategy
```

```python
data = apple.Close.loc[(apple.index>='1991-01-01')]
data = data.to_frame()
data
```

```python
sma_s = 50
sma_l = 100
```

```python
data["sma_s"] = data.Close.rolling(sma_s).mean()
data["sma_l"] = data.Close.rolling(sma_l).mean()
data
```

```python
data.plot(figsize=(12,8), title="Apple - SMA{} | SMA{}".format(sma_s, sma_l), fontsize=12)
```

```python
data.loc["2020"].plot(figsize=(12,8), title="Apple - SMA{} | SMA{}".format(sma_s, sma_l), fontsize=12)
```

```python
data.dropna(inplace=True)
data
```

```python
data["position"] = np.where(data["sma_s"] > data["sma_l"], 1, -1)
data
```

```python
data.loc["2020",["sma_s", "sma_l", "position"]].plot(figsize=(12,8), title="Apple - SMA{} | SMA{}".format(sma_s, sma_l), fontsize=12, secondary_y ="position")
plt.show()
```

```python
# buy and hold
data["ret_b&h"] = np.log(data.Close.div(data.Close.shift(1)))
data
```

```python
data["strategy"] = data["ret_b&h"] * data.position.shift(1)
data
```

```python
data.dropna(inplace=True)
```

```python
data[["ret_b&h", "strategy"]].sum()
```

```python
# What if I invested $1?
data[["ret_b&h", "strategy"]].sum().apply(np.exp)
```

```python
# annual std
data[["ret_b&h", "strategy"]].std()*np.sqrt(252)
```

```python
data.Close.plot(figsize=(12,8))
plt.show()
```

## Strategy adjusted with long bias

```python
data["position2"] = np.where(data["sma_s"] > data["sma_l"], 1, 0)
data["strategy2"] = data["ret_b&h"] * data.position2.shift(1)
data
```

```python
data.dropna(inplace=True)
data
```

```python
data[["ret_b&h", "strategy2"]].sum()
```

```python
# What if I invested $1?
data[["ret_b&h", "strategy2"]].sum().apply(np.exp)
```

```python
# annual std
data[["ret_b&h", "strategy2"]].std()*np.sqrt(252)
```

```python
## Strategy 2 is better than strategy 1. Less risk & similar or higher return
```
