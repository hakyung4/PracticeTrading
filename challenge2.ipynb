### Take the 5 stocks that I picked and plot the moving averages and graph it

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
TaiwanSemi = yf.download("TSM")
Costco = yf.download("COST")
Nike = yf.download("NKE")
PG = yf.download("PG")
Nvidia = yf.download("NVDA")
```

```python
TaiwanSemi = TaiwanSemi.Close.to_frame()
TaiwanSemi
```

```python
TaiwanSemi.plot(figsize=(15,8), fontsize=15)
plt.legend(loc="upper left", fontsize=15)
plt.show()
```

```python
TaiwanSemi["SMA50"] = TaiwanSemi.rolling(window=50, min_periods=50).mean()
```

```python
TaiwanSemi.plot(figsize=(15,8), fontsize=15)
plt.legend(loc="upper left", fontsize=15)
plt.show()
```

```python
Costco = Costco.Close.to_frame()
Costco
```

```python
Costco["SMA50"] = Costco.rolling(window=50, min_periods=50).mean()
```

```python
Costco.plot(figsize=(15,8), fontsize=15)
plt.legend(loc="upper left", fontsize=15)
plt.show()
```
